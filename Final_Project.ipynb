```python
df2 = pd.read_excel("Census_Data.xlsx")
df2
```

```python
df2 = df2.rename({'native-country':'country'}, axis='columns')
df2
```

```python
df2["country"] = df2["country"].apply(lambda x: x.replace("Dominican-Republic", "Dominican Republic"))
df2["country"] = df2["country"].apply(lambda x: x.replace("El-Salvador", "El Salvador"))
df2["country"] = df2["country"].apply(lambda x: x.replace("England", "United Kingdom"))
df2["country"] = df2["country"].apply(lambda x: x.replace("Holand-Netherlands", "Netherlands"))
df2["country"] = df2["country"].apply(lambda x: x.replace("Hong", "Hong Kong"))
df2["country"] = df2["country"].apply(lambda x: x.replace("Scotland", "United Kingdom"))
df2["country"] = df2["country"].apply(lambda x: x.replace("South", "South Korea"))
df2["country"] = df2["country"].apply(lambda x: x.replace("United-States", "United States"))
df2["workclass"] = df2["workclass"].apply(lambda x: x.replace("?", "Unknown"))
df2["occupation"] = df2["occupation"].apply(lambda x: x.replace("?", "Unknown"))
```

```python
df2.drop(['education-num', 'relationship', 'capital-gain', 'capital-loss', 'workclass', 'fnlwgt'], axis=1, inplace=True)
df2
```

```python
# Export transformed file to excel
df2.to_excel('clean_census_df.xlsx', index=False)
```

```python
# join the two tables
combined_df = pd.merge(df2, df, on="country", how="left")
combined_df
```

```python
list(combined_df)
```

```python
drop_df = combined_df.dropna()
drop_df
```
